```python
%matplotlib nbagg
import argparse
import importlib
import models
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from open3d import *
```

```python
def plot_pcd(ax, pcd):
    ax.scatter(pcd[:, 0], pcd[:, 1], pcd[:, 2], zdir='y', c=pcd[:, 0], s=0.5, cmap='Reds', vmin=-1, vmax=0.5)
    ax.set_axis_off()
    ax.set_xlim(-0.3, 0.3)
    ax.set_ylim(-0.3, 0.3)
    ax.set_zlim(-0.3, 0.3)
```

```python
pcd = read_point_cloud('/code/demo_data/car.pcd')
pcd_points = np.array(pcd.points)
pcd_downsample = pcd_points[::8]
```

```python
fig, axes = plt.subplots()
axes= plt.axes(projection='3d')
```

```python
plt.show()
plot_pcd(axes, pcd_points)
```

```python
fig2, axes2 = plt.subplots()
axes2= plt.axes(projection='3d')
```

```python
plt.show()
plot_pcd(axes2, pcd_downsample)
```

```python
def generate_3d():
    """Generate a 3D random rotation matrix.
    Returns:
        np.matrix: A 3D rotation matrix.
    """
    x1, x2, x3 = np.random.rand(3)
    R = np.matrix([[np.cos(2 * np.pi * x1), np.sin(2 * np.pi * x1), 0],
                   [-np.sin(2 * np.pi * x1), np.cos(2 * np.pi * x1), 0],
                   [0, 0, 1]])
    v = np.matrix([[np.cos(2 * np.pi * x2) * np.sqrt(x3)],
                   [np.sin(2 * np.pi * x2) * np.sqrt(x3)],
                   [np.sqrt(1 - x3)]])
    H = np.eye(3) - 2 * v * v.T
    M = -H * R
    return M
```

```python
M = generate_3d()
```

```python
pcd_points.shape
```

```python
import cv2
pcd_points_rotated= cv2.transform(np.reshape(pcd_points, (1511,1,3)), M)
```

```python
plt.show()
plot_pcd(axes5,pcd_points_rotated.reshape(1511,3))
```

```python
fig5, axes6 = plt.subplots()
axes6 = plt.axes(projection='3d')
```

```python
plt.show()
plot_pcd(axes6,complete_rotated)
```

```python
def generate_2d():
    """Generate a 2D random rotation matrix.
    Returns:
        np.matrix: A 2D rotation matrix.
    """
    x = np.random.random()
    M = np.matrix([[np.cos(2 * np.pi * x), -np.sin(2 * np.pi * x)],
                   [np.sin(2 * np.pi * x), np.cos(2 * np.pi * x)]])
    return M
```

```python
M_2 = generate_2d()
```

```python
M3_z = np.eye(3)
```
